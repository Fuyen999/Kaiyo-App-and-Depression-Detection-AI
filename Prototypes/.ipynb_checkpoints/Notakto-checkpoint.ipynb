```python
import random

#stores data in individual boards form
size = 3
boards = {'A':[i for i in range(size**2)], 'B':[i for i in range(size**2)], 'C':[i for i in range(size**2)]}
player = (2, 1)

def print_board():
    #convert boards into rows
    current_board = {0:[], 1:[], 2:[], 3:[]}
    for i in boards:
        current_board[0].append(i)
        for k in range(1,4):
            current_board[k].append([])
        for j in range(len(boards[i])):
            current_board[(j//size)+1][-1].append(str(boards[i][j]))
    
    #print out the rows properly
    print('      '.join(current_board[0]))
    for i in range(1,4):
        for j in range(len(current_board[i])):
            current_board[i][j] = ' '.join(current_board[i][j])
        print('  '.join(current_board[i]))
    
def choose_even_board():
    even_boards = []
    for i in boards:
        number_of_X = boards[i].count('X')
        if number_of_X % 2 == 0:
            even_boards.append(i)
    if even_boards != []:
        board = random.choice(even_boards)
        return board
    return False
    
def check_if_straight_line_is_formed(linelist, num):
    linelist[num//size] += 1
    linelist[(num%size) + size] += 1
    if num//size == num%size:
        linelist[-2] += 1
    if num//size + num%size == size-1:
        linelist[-1] += 1
    return linelist
    
#def numbers_that_can_form_straight_line(linelist, numboard):
def numbers_that_can_form_straight_line(board, linelist):
    points = []
    straight_points = []
    for j in board:
        if j == 'X':
            continue
        else:
            variable_linelist = linelist.copy()
            variable_linelist = check_if_straight_line_is_formed(variable_linelist, j)
            if size in variable_linelist:
                points.append(j)
    return points

def check_if_centre_empty():
    flag = False
    board = []
    for i in boards:
        if boards[i][4] != 'X':
            flag = True
            board.append(i)
    return flag, board

def create_linelist_from_board(board):
    linelist = [0 for i in range(size*2+2)]
    for i in range(len(board)):
        if board[i] == 'X':
            linelist = check_if_straight_line_is_formed(linelist, i)
    return linelist

def number_that_wont_overload_board(numboard):
    free_num_list = []
    forbidden_list = []
    straight_line_list = []
    for i in range(len(numboard)):
        variable_board = numboard.copy()
        if variable_board[i] == 'X':
            continue
        else:
            variable_board[i] = 'X'
        
        variable_linelist = create_linelist_from_board(numboard).copy()
        new_linelist = check_if_straight_line_is_formed(variable_linelist, i)
        if size in new_linelist:
            straight_line_list.append(i)
            continue 
        
        #variable_board[4] = 4
        #newest_linelist = create_linelist_from_board(variable_board)
        next_num_list = numbers_that_can_form_straight_line(variable_board, new_linelist)
        if next_num_list == []:
            free_num_list.append(i)
        if next_num_list == [i for i in variable_board if i != 'X']:
            forbidden_list.append(i)
    
    allowed_list = [i for i in numboard if i != 'X' and i not in forbidden_list and i not in straight_line_list]
    if free_num_list != []:
        return free_num_list
    
    elif allowed_list != []:
        for_iterate_purpose = allowed_list.copy()
        for num in for_iterate_purpose:
            variable_board = numboard.copy()
            variable_board[num] = 'X'
            if variable_board.count('X') == 3:
                next_allowed_list = number_that_wont_overload_board(variable_board)
                new_straight_line_list = []
                for i in [j for j in variable_board if j != 'X']:
                    variable_variable_linelist = create_linelist_from_board(variable_board).copy()
                    newest_linelist = check_if_straight_line_is_formed(variable_variable_linelist, i)
                    if size in newest_linelist:
                        new_straight_line_list.append(i)
                if next_allowed_list != [i for i in variable_board if i != 'X' and i not in new_straight_line_list]:
                    allowed_list.remove(num)
        
        return allowed_list
    
    return False

def check_how_many_Xs(count):
    three_X_board = []
    for i in boards:
        if boards[i].count('X') == count:
            three_X_board.append(i)
    if three_X_board != []:
        board = random.choice(three_X_board)
        return board
    return False
        
def three_boards():
    flag = False
    for i in boards:
        board = i
        linelist = boardlist[boardlist_push_reference.index(board)]
        straight_line_numlist = numbers_that_can_form_straight_line(boards[board], linelist)
        if straight_line_numlist != []:
            num = random.choice(straight_line_numlist)
            flag = True
            break
    
    if flag == False:
        board = choose_even_board()
        if boards[board].count('X') == 0:
            num = 4
        else:
            num = random.choice(number_that_wont_overload_board(boards[board]))
        
    return board, str(num)

def two_boards():
    centre_empty_bool, empty_board_list = check_if_centre_empty()
    three_X_board = check_how_many_Xs(3)
    two_X_board = check_how_many_Xs(2)
    
    flag = False
    if two_X_board != False:
        board = two_X_board
        linelist = boardlist[boardlist_push_reference.index(board)]
        straight_line_numlist = numbers_that_can_form_straight_line(boards[board], linelist)
        if board in empty_board_list:
            flag = True
            if straight_line_numlist != []:
                num = random.choice(straight_line_numlist)
            else:
                num = random.choice(number_that_wont_overload_board(boards[board]))
    
    if flag == False:
        if centre_empty_bool == True:
            even_board = choose_even_board()
            if even_board in empty_board_list:
                board = even_board
            else:
                board = random.choice(empty_board_list)
            num = 4
            
        elif three_X_board != False:
            board = three_X_board
            num = random.choice(number_that_wont_overload_board(boards[board]))
    
        else:
            board = choose_even_board()
            linelist = boardlist[boardlist_push_reference.index(board)]
            straight_line_numlist = numbers_that_can_form_straight_line(boards[board], linelist)
            num = random.choice(straight_line_numlist)
    
    return board, str(num)

def one_board():
    centre_empty_bool, empty_board_list = check_if_centre_empty()
    board = boardlist_push_reference[0]
    suitable_num_list = number_that_wont_overload_board(boards[board])

    if centre_empty_bool == True:
        num = 4

    elif suitable_num_list != False:
        num = random.choice(suitable_num_list)
        
    else:
        linelist = boardlist[boardlist_push_reference.index(board)]
        straight_line_numlist = numbers_that_can_form_straight_line(boards[board], linelist)
        num = random.choice([i for i in boards[board] if i != 'X' and i not in straight_line_numlist])
        
    return board, str(num)
        
def best_move():
    if len(boards) == 3:
        return three_boards()
    elif len(boards) == 2:
        return two_boards()
    else:
        return one_board()
    
round = 1
boardlist = [[0 for i in range(size*2+2)] for i in range(len(boards))]
boardlist_push_reference = ['A', 'B', 'C']

print_board()
while len(boards) != 0:
    #input player's move
    if round % 2 == 0:
        move = list(input('Player 2: '))
    else:
        move = best_move()
        movestring = ''.join(move)
        print(f'Player 1: {movestring}')

    #Check if input is valid
    flag = False
    if len(move) == 2:
        board, num = move
        if num.isalpha() == False and board.isupper() == True:
            num = int(num)
            if num < size**2 and board in boards:
                if boards[board][num] != 'X':
                    boards[board][num] = 'X'
                    round += 1
                    flag = True
    
    if flag == False:
        print('Invalid move, please input again')
        continue
        
    linelist = boardlist[boardlist_push_reference.index(board)]
    linelist = check_if_straight_line_is_formed(linelist, num)
    
    #remove dead board
    if size in linelist:
        boardlist.remove(linelist)
        boardlist_push_reference.remove(board)
        del boards[board]
    
    #print board for next round
    if len(boards) != 0:
        print_board()
    
print(f'Player {player[round%2]} wins game')
```
